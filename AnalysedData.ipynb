```python
def load_text(filename):
    with open(filename,encoding='utf-8') as f:
        data=f.read()
    return data
```

```python
data=load_text("FinalData.txt")
data
```

```python
chunks= data.split("\n\n")
chunks[0]
```

```python
chunks=[c for c in chunks if len(c)>5]
```

```python
def parse(chunk):
    chunk=chunk.strip()
    sepchunk=chunk.split("\n")
    user_name=sepchunk[0]
    posts=int(sepchunk[1].split(' posts')[0].replace(",","").replace(" post",""))
    followers=int(sepchunk[2].split(' followers')[0].replace(",","").replace(".","").replace("K","000").replace("M","000000"))
    following=int(sepchunk[3].split(' following')[0].replace(",","").replace(".","").replace("K","000").replace("M","000000"))
    bio=sepchunk[4:]

    final_chunk={'User Name-': user_name,
    'Posts-':posts,
    'Followers-':followers,
    'Following-':following,
    'Bio-':bio}

    return final_chunk
    
```

```python
all_chunks=[]
count=0
for chunk in chunks :
    count+=1
    all_chunks.append(parse(chunk))
    print(parse(chunk))
    print('\n')

count
```

```python
import json 

def dump_data(data):
    dumped_data=json.dumps(data ,indent=3)
    with open("insta_data.json","w") as f:
        f.write(dumped_data)
    
    return dumped_data
```

```python
print(dump_data(all_chunks))
```


# Max Posts


```python
max_post=0
for chunk in all_chunks:
    max_post=max(max_post,chunk['Posts-'])

print(f'Max Posts is {max_post} \n' )

for chunk in all_chunks:
    if chunk['Posts-']==max_post :
        print(chunk)    
```
